```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from web_site_content import fetch_website_contents
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI()
```

```python
ed = fetch_website_contents("https://edwarddonner.com")

system_prompt = """
You are a cheerful, upbeat assistant whose primary goal is to help the user understand content. 
Prioritize accuracy and faithfulness to the source over entertainment. 
Keep a friendly tone, but avoid humor when the topic is sensitive or technical. 
When summarizing, be concise, avoid speculation, and do not invent facts. 
If content is missing, ask for the page text or URL. Write outputs in the language the user requests.
"""

user_prompt = """
This is a website. Read the content below.
Task:
(1) Produce a concise summary in English capturing key points, entities, numbers, and dates. 
(2) Provide a faithful Japanese translation of that summary.
Constraints: Prioritize accuracy; preserve terminology; do not add information not present in the text. 
If the provided text is insufficient, say what is missing and ask for the rest.
"""
```

```python
def prompt_put(website):
    return [
    {"role" : "system", "content" : system_prompt},
    {"role" : "user", "content" : user_prompt + website}
    ]
```

```python
prompt_put(ed)
```

```python
def summarize(url):
    website = fetch_website_contents(url)
    response = openai.chat.completions.create(
        model = "gpt-4.1-mini",
        messages = prompt_put(website)
    )
    return response.choices[0].message.content
```

```python
def display_summarize(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summarize("https://edwarddonner.com")
```
